```python
import algos
import utils
import numpy as np
```

```python
nb_tasks = 10
#instance, erreur = utils.generateInstance(nb_tasks, v_generator=utils.poisson_generator)
colors = np.random.rand(nb_tasks, 3)
```

```python
#PART 1
instance, erreur = utils.generateInstance(nb_tasks, x_generator=utils.pareto_generator)
for task in instance:
    print('id ', task.id, 'time_estimated ', task.time_estimated)
#Predict
solution = algos.PREDICT(instance)
obj = utils.evaluateObjective(instance, solution)
utils.plotGant(solution, nb_tasks, colors, obj, filename = 'ParetoPREDICT.png')
#SPT
solution = algos.SPT(instance)
obj = utils.evaluateObjective(instance, solution)
utils.plotGant(solution, nb_tasks, colors, obj, filename = 'ParetoSPT.png')
#Round Robin
solution = algos.ROUND_ROBIN(instance)
obj = utils.evaluateObjective(instance, solution)
utils.plotGant(solution, nb_tasks, colors, obj, filename = 'ParetoRR.png')
```
